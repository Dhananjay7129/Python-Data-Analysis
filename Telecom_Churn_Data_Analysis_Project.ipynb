```python
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"


df ["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = "Churn", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by Churn")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame (Replace this with your actual dataset)
# df = pd.read_csv("your_data.csv")

# Compute churn percentage within each SeniorCitizen group
grouped = df.groupby("SeniorCitizen")["Churn"].value_counts(normalize=True).unstack() * 100

# Define color palette
colors = ["#1f77b4", "#ff7f0e"]  # Blue for No, Orange for Yes

# Plot
fig, ax = plt.subplots(figsize=(6, 5))
grouped.plot(kind="bar", stacked=True, color=colors, ax=ax, edgecolor="black", width=0.6)

# Add percentage labels
for i, (index, row) in enumerate(grouped.iterrows()):
    cumulative = 0
    for churn_category, value in row.items():
        if value > 5:  # Only label if % is significant
            ax.text(i, cumulative + value / 2, f"{value:.1f}%",
                    ha="center", va="center", color="white", fontsize=12, fontweight="bold")
        cumulative += value

# Styling
plt.xlabel("Senior Citizen", fontsize=13, fontweight="bold")
plt.ylabel("Percentage (%)", fontsize=13, fontweight="bold")
plt.title("Churn Distribution by Senior Citizen", fontsize=14, fontweight="bold")
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

# Increase y-axis limit slightly above 100%
plt.ylim(0, 105)

# Add a grid for better readability
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Adjust legend
plt.legend(title="Churn", labels=["No", "Yes"], fontsize=12, title_fontsize=12)

# Show the plot
plt.show()

```

```python
# comparitive a greater percentage of people in senior citizen category have churned
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplots in a 3x3 grid
fig, axes = plt.subplots(3, 3, figsize=(12, 10))

# Loop through each column and create a count plot
for col, ax in zip(columns, axes.flatten()):
    sns.countplot(data=df, x=col, ax=ax, hue = df["Churn"])
    ax.set_title("count of " + col)  # Set title
    ax.set_xlabel(" ")  # Remove x-axis label for clarity
    ax.set_ylabel("Count")
    ax.tick_params(axis='x')
plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
# Customers without additional services like Online Security, Tech Support, and Device Protection appear to have higher churn rates. Fiber optic users also show a noticeable churn rate compared to DSL. The trend suggests that lack of value-added services might contribute to customer churn.
```
